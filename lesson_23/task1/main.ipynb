```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

### Эксперимент 1

```python
capasity = np.loadtxt("exp1.txt")
size = np.arange(len(capasity)) + 1
breakPoints = capasity == size
breakPoints = breakPoints.nonzero()[0] + 2
MNK = linregress(size[breakPoints], capasity[breakPoints])
```

```python
fig, ax = plt.subplots(figsize = (16,9), dpi = 200)
xAx = np.linspace(0, len(capasity), 2000)

plt.plot(capasity, label = 'Экспериментальные данные')
plt.scatter(size[breakPoints], 
            capasity[breakPoints], 
            color = 'red', 
            label = 'Выделена новая память')
plt.plot(xAx, 
         MNK.slope * xAx + MNK.intercept,
         label = r"Интерполяция $\alpha = {}$".format(MNK.slope))
plt.xlabel("Количество добавленных элементов")
plt.ylabel("Capasity")
plt.grid()
plt.xlim(0, len(capasity))
plt.ylim(0,capasity[-1]+5)
plt.legend()
plt.title("Эксперимент 1 - вектор изначально пуст");
```

Вывод - при переполнении capasity увеличивается вдвое:

```python
capasity[breakPoints] - size[breakPoints] * 2 <= 1e-6
```
